```python
df['Views'].fillna(df['Views'].mean(), inplace=True)
df['Likes'].fillna(df['Likes'].mean(), inplace=True)
df['Comments'].fillna(df['Comments'].mean(), inplace=True)
df['Stream'].fillna(df['Stream'].mean(), inplace=True)
```

```python
album_type_count=df["Album_type"].value_counts()
print(album_type_count)
colors = plt.get_cmap('tab20').colors
plt.figure(figsize=(8, 8))
plt.pie(album_type_count, labels=album_type_count.index, autopct='%1.1f%%', startangle=140, colors=colors[:len(album_type_count)])
plt.title('Albüm tip dağılımı')
plt.axis('equal') 
plt.show()
```

```python
#PLATFORMA GÖRE İLK 10 ŞARKI
```

```python
platform = df.groupby("Artist").agg({"Stream" : "mean", "Views" : "mean"}).sort_values(by = ["Stream", "Views"], ascending = [False, False]).head(10).reset_index()
platform
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(14, 12))
sns.barplot(x="Stream", y="Artist", data=platform, palette='viridis', ax=axes[0])
axes[0].set_title('Top 10 Artists by Average Stream')
axes[0].set_xlabel('Average Stream Count')
axes[0].set_ylabel('Artist')
sns.barplot(x="Views", y="Artist", data=platform, palette='magma', ax=axes[1])
axes[1].set_title('Top 10 Artists by Average Views')
axes[1].set_xlabel('Average Views Count')
axes[1].set_ylabel('Artist')
plt.tight_layout()
plt.show()
```

```python
top_songs = df.sort_values('Stream', ascending=False).head(10)
top_songs[['Track', 'Energy', 'Danceability', 'Valence']]
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))
viridis = sns.color_palette("viridis", 3)
sns.barplot(x='Track', y='Energy', data=top_songs, color=viridis[0], label='Energy')
sns.barplot(x='Track', y='Danceability', data=top_songs, color=viridis[1], label='Danceability', alpha=0.6)
sns.barplot(x='Track', y='Valence', data=top_songs, color=viridis[2], label='Valence', alpha=0.3)
plt.xticks(rotation=90, ha='right')
plt.legend()
plt.title('Top 10 Songs by Stream: Energy, Danceability, and Valence')
plt.xlabel('Track')
plt.ylabel('Values')
plt.show()
```
